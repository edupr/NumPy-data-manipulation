### Importar librerías

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7'],})
```

```python
df1
```

```python
df2
```

### concat()
concat normal, simpplemente pegar ambos DF como un stack:

```python
pd.concat([df1,df2])
```

Para corregir el índice:

```python
pd.concat([df1,df2],ignore_index=True)
```

concat con columnas (axis=1)

```python
pd.concat([df1,df2],axis=1)
```

### merge()

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

merge tradicional con columnas en común

```python
izq.merge(der)
```

merge especificando el común: la key

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

Cuando no hay un key común:

```python
izq.merge(der, on='key')
```

Con esto sí podemos hacer merge:

```python
izq.merge(der, left_on='key', right_on='key_2')
```

DF con NaN:

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
der
```

La key=k3 se elimina porque no hay match por el NaN:

```python
izq.merge(der, left_on='key', right_on='key_2')
```

### how  Me trae el key de la izquierda:

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

### how  Me trae el key de la derecha:

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

### join() - Index match

```python
izq = pd.DataFrame({'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2'])

der = pd.DataFrame({'C':['C0','C1','C2'],
 'D':['D0','D1','D2']},
 index=['k0','k2','k3'])
```

```python
izq
```

```python
der
```

a la izquierda le une la derecha:

```python
izq.join(der)
```

une lo común:

```python
izq.join(der, how='inner')
```

pone todo y donde no hay valores, pone NaN:

```python
izq.join(der,how='outer')
```
